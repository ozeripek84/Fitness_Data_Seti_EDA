# <p style="background-color:blue;font-family:newtimeroman;font-size:100%;color:white;text-align:center;border-radius:20px 20px;"><b>About the Dataset:</b></p>



- Age: The age of the individual (years).
- Gender: The gender of the individual (e.g., Male, Female).
- Weight (kg): The weight of the individual (kilograms).
- Height (m): The height of the individual (meters).
- Max_BPM: The maximum heart rate recorded during exercise.
- Avg_BPM: The average heart rate during a workout session per minute.
- Resting_BPM: The resting heart rate per minute.
- Session_Duration (hours): The duration of the exercise session (hours).
- Calories_Burned: The total calories burned during a workout session.
- Workout_Type: The type of exercise performed (e.g., Cardio, Strength, Yoga).
- Fat_Percentage: Body fat percentage.
- Water_Intake (liters): Water intake during or after exercise, measured in liters.
- Workout_Frequency (days/week): The number of days the individual works out per week.
- Experience_Level: The fitness experience level (e.g., Beginner, Intermediate, Advanced).
- BMI: Body Mass Index, calculated as weight (kg) / height (m)^2.

### We identified and corrected erroneous data entries.

```python
df["Workout_Type"] = df["Workout_Type"].replace({"\\nStrength": "Strength", "\\tCardio":"Cardio","\\tYoga": "Yoga"})
df["Workout_Type"].value_counts()
```

```python
# df["Workout_Type"] = df["Workout_Type"].str.replace("\\nt", "")  ==>>> Second Way.
```

```python
# There were incorrect values in the BMI calculation. We recalculated and reassigned them accordingly.

df['BMI'] =round(df['Weight (kg)'] / df['Height (m)'] ** 2, 2)
df['BMI'].head(10)
```

```python
# We classified individuals into BMI (Body Mass Index) categories and created a new column accordingly.

def categorize_bmi(bmi):
    if bmi <= 18.5:
        return 'Underweight'
    elif 18.5 < bmi <= 24.99:
        return 'Normal weight'
    elif 25.0 <= bmi <= 29.99:
        return 'Overweight'
    elif 30.0 <= bmi <= 34.99:
        return 'Obesity Class 1'
    elif 35.0 <= bmi <= 39.99:
        return 'Obesity Class 2'
    else:
        return 'Obesity Class 3'
```

```python
df['BMI_Category'] = df['BMI'].transform(categorize_bmi)
df['BMI_Category']
```

```python
# We corrected the incorrect data in the Max_BPM column.
df["Max_BPM"] = df["Max_BPM"].str.strip("\\tn")
```

```python
# We changed the data type of `max_bpm` from object to float.
df["Max_BPM"] = df["Max_BPM"].astype("float64")
```

```python
# Then, we filled the missing values with the median.

df["Avg_BPM"].fillna(df["Avg_BPM"].median(), inplace=True)
df["Max_BPM"].fillna(df["Max_BPM"].median(), inplace=True)
```

```python
# We created a new column to determine workout intensities.

df["Workout_Intensity"] = round(df["Avg_BPM"] / df["Max_BPM"],2)
df["Workout_Intensity"]
```
